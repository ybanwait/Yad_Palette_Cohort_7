# Elbow Method

##### Using elbow method to find out optimal number of clusters

```python
# One-hot encode the 'Variety' column
data_encoded = pd.get_dummies(df_oats_feature, columns=['Variety'])

# Select numerical features including the encoded columns
features = data_encoded[['Farms', 'Acres', 'Yield_Float'] + [col for col in data_encoded.columns if 'Variety_' in col]]

# Normalize the features
scaler = StandardScaler()
normalized_features = scaler.fit_transform(features)

# Apply k-means for a range of cluster numbers to find the optimal one using the elbow method
inertia_adjusted = []
k_range_adjusted = range(1, 8)  # Adjusting the range for fewer clusters

for k in k_range_adjusted:
    kmeans = KMeans(n_clusters=k, max_iter=100, random_state=42)  # Limited iterations for speed
    kmeans.fit(normalized_features)
    inertia_adjusted.append(kmeans.inertia_)

# Plot the elbow method
plt.figure(figsize=(8, 6))
plt.plot(k_range_adjusted, inertia_adjusted, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Sum of Squared Distances)')
plt.title('Elbow Method for Optimal k (Adjusted)')
plt.grid(True)
plt.show()
```

```python
# Define the range of clusters for silhouette score evaluation
k_range_adjusted = range(2, 8)  # Start from 2 clusters since k=1 is not valid for silhouette scores
silhouette_scores = []

# Calculate silhouette scores
for k in k_range_adjusted:
    kmeans = KMeans(n_clusters=k, max_iter=100, random_state=42)
    labels = kmeans.fit_predict(normalized_features)
    score = silhouette_score(normalized_features, labels)
    silhouette_scores.append(score)

# Plot silhouette scores
plt.figure(figsize=(8, 6))
plt.plot(k_range_adjusted, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different Cluster Counts')
plt.grid(True)
plt.show()
```

```python
#From Silhouette score we can confirm the optimal number of clusters is 3.
#Apply k-means clustering with 3 clusters
kmeans_3_clusters = KMeans(n_clusters=3, max_iter=100, random_state=42)
cluster_labels = kmeans_3_clusters.fit_predict(normalized_features)

# Step 5: Add the cluster labels to the original data
df_oats_feature['Cluster'] = cluster_labels

# Step 6: Analyze and summarize each cluster
cluster_summary = df_oats_feature.groupby('Cluster')[['Farms', 'Acres', 'Yield_Float']].mean()
print(cluster_summary)
```
